```python
#Leitura do dataset
df = pd.read_csv('2016.csv', encoding='latin1', delimiter=';')
```

```python
#Alterando virgulas por pontos e transformando coluna em inteiro
df['VALOR_REEMBOLSADO'] = df['VALOR_REEMBOLSADO'].str.replace(',', '.')
df['VALOR_REEMBOLSADO'] = pd.to_numeric(df['VALOR_REEMBOLSADO'])
```

```python
#Selecionando linhas de um dataset com base em uma condição
df[df['DATA'] == '20/07/5017']
```

```python
#Selecionando linhas em que haja 5017 nos ultimos quatro caracteres e colocando um novo valor
df.loc[df['DATA'].str[-4:] == '5017','DATA'] = '20/07/2017'
```

```python
#Alterando datas estão com 2106 para 2016
df.loc[df['DATA'].str[-4:] == '2106','DATA'] = '2016'
```

```python
#Alterando datas estão com 0216 para 2016
df.loc[df['DATA'].str[-4:] == '0216','DATA'] = '2016'
```

```python
#Alterando datas estão com 3016 para 2016
df.loc[df['DATA'].str[-4:] == '3016','DATA'] = '2016'
```

```python
#Selecionando dados inválidos com base em multiplas condições
df.loc[(df['DATA'].str[-4:] != '2016') & (df['DATA'].str[-4:] != '2017') & (df['DATA'].str[-4:] != '2015')]
```

```python
#Transformando a coluna de data em datetime, considerando datas invalidas como nulas
df['DATA'] = pd.to_datetime(df['DATA'], errors='coerce')
```

```python
#Selecionando valores unicos da coluna Tipo de Despesa
df['TIPO_DESPESA'].unique()
```

```python
#Mapeando novos valores para a coluna Tipos de Despesa
df['TIPO_DESPESA'] = df['TIPO_DESPESA'].map({'Aluguel de imóveis para escritório político, compreendendo despesas concernentes a eles.': 'Aluguel',
                                             'Aquisição de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos. ': 'Despesas gerais',
                                             'Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar': 'Contratações',
                                             'Locomoção, hospedagem, alimentação, combustíveis e lubrificantes': 'Locomoção, hospedagem e alimentação',
                                             'Passagens aéreas, aquáticas e terrestres nacionais': 'Passagens',
                                             'Divulgação da atividade parlamentar': 'Divulgação',
                                             'Serviços de Segurança Privada': 'Segurança'})
```
